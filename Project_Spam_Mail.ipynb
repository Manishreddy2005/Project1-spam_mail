```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv("Mail_Data.csv")
data
```

```python
# To know the names of columns
print(data.columns)
```

```python
# data.rename(columns={'Message': 'text'}, inplace=True)
# data[data['Category'] == 'spam']
# print(data['Category'].unique())
```

```python
# Split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data["Message"],data["Category"], test_size=0.2, random_state=10)
```

```python
# Import Count vectorizer from feature extraction
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words="english")
vect.fit(X_train)
```

```python
X_train_df = vect.transform(X_train)
X_test_df = vect.transform(X_test)
X_train_df
```

```python
# Ploting the Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
plt.figure(figsize=(150,100))
model = DecisionTreeClassifier()
model.fit(X_train_df,y_train)
pred = model.predict(X_test_df)
plot_tree(model, filled=True, feature_names=vect.get_feature_names_out())
```

```python
#import Multinomail naive bayes from naive bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
model = MultinomialNB(alpha=1.8)
model.fit(X_train_df,y_train)
pred = model.predict(X_test_df)
```

```python
accuracy=accuracy_score(y_test,pred)
print("Accuracy:",accuracy*100)
```

```python
# Evaluate the model performeance
print(classification_report(y_test, pred , target_names = ["Not Spam", "Spam"]))
```
